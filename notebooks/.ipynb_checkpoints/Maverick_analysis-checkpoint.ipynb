```python
import os 
os.chdir('../')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from config import *
from utilities import calculate_amounts_agg
```

```python
deposits_df = pd.read_csv(MAVERICK_TOKEN_DEPOSITS_FILENAME)
withdrawals_df = pd.read_csv(MAVERICK_TOKEN_WITHDRAWALS_FILENAME)
reprice_df = pd.read_csv(MAVERICK_TOKEN_REPRICE_FILENAME)
transfers_df = pd.read_csv(MAVERICK_TOKEN_TRANSFER_FILENAME)

deposits_df.shape, withdrawals_df.shape, reprice_df.shape, transfers_df.shape
```

```python
transfers_df.tail()
```

# Preprocessing

```python
deposits_df['block_date'] = pd.to_datetime(deposits_df['block_date'])
deposits_df['date'] = pd.to_datetime(deposits_df['block_date']).dt.date
deposits_df['amount'] = deposits_df['amount'].astype(float)/1e18
deposits_df['swETHMinted'] = deposits_df['swETHMinted'].astype(float)/1e18
deposits_df['newTotalETHDeposited'] = deposits_df['newTotalETHDeposited'].astype(float)/1e18
deposits_df = deposits_df[~pd.isnull(deposits_df.swETHMinted)]

withdrawals_df['block_date'] = pd.to_datetime(deposits_df['block_date'])
withdrawals_df['date'] = pd.to_datetime(deposits_df['block_date']).dt.date
#withdrawals_df['swETHBurned'] = withdrawals_df['swETHBurned'].astype(float)/1e18
#withdrawals_df['ethReturned'] = withdrawals_df['ethReturned'].astype(float)/1e18


reprice_df['block_date'] = pd.to_datetime(reprice_df['block_date'])
reprice_df['date'] = pd.to_datetime(reprice_df['block_date']).dt.date
reprice_df['amount'] = withdrawals_df['amount'].astype(float)/1e18
reprice_df['swETHMinted'] = withdrawals_df['swETHMinted'].astype(float)/1e18

transfers_df['block_date'] = pd.to_datetime(transfers_df['block_date'])
transfers_df['date'] = pd.to_datetime(transfers_df['block_date']).dt.date
transfers_df['value'] = transfers_df['value'].astype(float)/1e18

transfers_df = transfers_df[~pd.isnull(transfers_df.value)][['date','value']]
```

# Volume

```python
daily_volume = transfers_df.groupby('date')['value'].sum()

cumulative_volume = daily_volume.cumsum()

fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.bar(daily_volume.index, daily_volume.values, color='lightblue', label='Daily Volume')
ax1.set_ylabel('Daily Volume')
ax1.set_xlabel('Date')

ax2 = ax1.twinx()

ax2.plot(cumulative_volume.index, cumulative_volume.values, color='orange', marker='o', label='Cumulative Volume')
ax2.set_ylabel('Cumulative Volume')

ax1.set_title('Daily and Cumulative Volume Over Time')
ax1.tick_params(axis='x', rotation=45)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
